```python
from polydiff.dataloader import DiffusionSmilesDataset, create_diffusion_dataloader
import torch

from transformers import AutoTokenizer, AutoModelForCausalLM
tokenizer = AutoTokenizer.from_pretrained("seyonec/ChemBERTa-zinc-base-v1")
```

```python
data_file = "/Users/wang-work/polydiffusion/data/smiles.txt"
data = []
with open(data_file, "r") as f:
    for line in f:
        line = line.strip()
        if line:
            data.append(line)
dataset = DiffusionSmilesDataset(data, tokenizer=tokenizer)
dataloader = create_diffusion_dataloader(
    data,
    batch_size=4,
)
```
